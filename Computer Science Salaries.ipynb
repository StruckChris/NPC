```python
import pandas as pd
```

```python
SalaryDF = pd.read_csv('SalaryDFwithPop.csv')
SalaryDF.drop('Unnamed: 0',inplace=True,axis=1)
```

```python
SalaryDF.head()
```

```python
SalaryDF2 = SalaryDF.where(SalaryDF['Occupational Area'] == 'Computer &amp; Mathematical').dropna(subset = 'Occupational Area')
SalaryDF2.drop_duplicates(subset='State',inplace=True)
SalaryDF2.drop(['County','2022 Population Est'],inplace=True,axis=1)
SalaryDF2['Occupational Area'] = ['Computer & Mathematical' for x in SalaryDF2['Occupational Area']]
```

```python
SalaryDF2
```

```python
SalaryDF2.to_csv('ComputerScience.csv')
```
